```python
url1 = "https://webscraper.io/test-sites/e-commerce/static"
response = requests.get(url1)
print(response.status_code)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup.title.text)
```

```python
products = soup.find_all("a", class_="title")

for p in products:
    print(p.text.strip())
```

```python
descriptions = soup.find_all("p", class_="description")

for d in descriptions:
    print(d.text.strip())
```
