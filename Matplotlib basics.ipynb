```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Matplotlib - The big library which contains pyplot
#Pyplot - The part which draws the plot
```

```python
x_axis = [1, 2, 3, 4, 5]
y_axis = [2, 3, 5, 7, 11]
plt.plot(x_axis,y_axis)
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.plot(x_axis,y_axis, marker= 'd', linestyle= '--', color= 'purple', label= 'Sales', linewidth= 2)
plt.title("Matplotlib Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.bar(x_axis,y_axis, linestyle= '--', color= 'skyblue', label= 'Sales', linewidth= 2)
plt.title("Bar Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
colors = ('grey', 'skyblue', 'lightgreen', 'pink', 'purple')
plt.figure(figsize = (8,7))
plt.bar(x_axis, y_axis, color = colors, width= 0.5)
plt.xlabel('X-axis')
plt.title("Bar Chart")
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.hist(y_axis, bins=6, color= 'lightgreen', edgecolor= 'darkgreen', alpha= 0.8)
plt.title("Histogram Chart")
```

```python
plt.scatter(x_axis, y_axis, color = 'orange', marker= '*', s = 100)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

```

```python
plt.boxplot(y_axis, vert= True, patch_artist= True)
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.plot(data['Year_of_Release'], data['Global_Sales'], marker= '*', color= 'skyblue', linewidth= 2, linestyle= '--')
plt.title("Global Sales by Year of Release")
plt.xlabel("Year of Release")
plt.ylabel("Global Sales(in millions)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.bar(data['Release_Year'], data['Global_Sales'], color= 'skyblue', width=2)
plt.title("Global Sales based on Year Release")
plt.xlabel("Release Year")
plt.ylabel("Global Sales")
plt.grid(True)
plt.show()
```

```python
top_games = data.groupby("Name")["Global_Sales"].sum().sort_values(ascending=False).head(10)
colors = ['pink', 'purple', 'skyblue', 'teal']
plt.bar(top_games.index, top_games.values, color= colors)
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 PS4 Games by Sales")
plt.show()
```
